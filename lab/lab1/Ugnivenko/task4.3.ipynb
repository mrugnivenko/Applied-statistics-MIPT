# Ответы 

- Недостаток этого способа состоит в том, что не учли поправку на множественную проверку гипотез. Чтобы улучшить данный способ, надо ввести эту поправку.
- 100 фотографий - довольно много, поэтому стоит ввести самую мощную поправку - Holm. 
- Он контролирует метрику качества FWER. FWER - это групповая вероятность ошибки первого рода. Метод Holm  гарантирует, что вероятность совершения хотя бы одной ошибки первого рода не превосходит $\alpha$.
- Данный метод имеет низкую мощность - это следствие контроля FWER на уровне значимости $\alpha$.
- Мощность станет ниже.
- Тогда в данном случае, лучше рассматривать метрику качества FDR - ожидаемая доля ложных отклонений гипотез. Так как теперь мы знаем, что статистики - независимы, то можно использовать поправку Бенджамини-Хохберга.
- Мощность увеличилась, так как смягчилось условия на количество ошибок первого рода.
